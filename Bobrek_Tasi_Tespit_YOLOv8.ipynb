```python
from ultralytics import YOLO
import os
import random
import cv2
from google.colab.patches import cv2_imshow


model = YOLO('/content/best (1).pt')

# Test resimlerinin yolu
test_yolu = '/content/bobrek_verisi/test/images'

if os.path.exists(test_yolu):
    # Rastgele bir resim seç
    rastgele_resim = random.choice(os.listdir(test_yolu))
    resim_adresi = os.path.join(test_yolu, rastgele_resim)

    # Tahmin yap
    print(f" Test ediliyor: {rastgele_resim}")
    sonuclar = model.predict(source=resim_adresi, conf=0.25)

    # Sonucu ekrana bas
    for r in sonuclar:
        im_array = r.plot()
        cv2_imshow(im_array)
else:
    print("⚠️ 'bobrek_verisi' klasörü bulunamadı. Lütfen önce veri indirme adımını çalıştır.")
```
